```python
import pandas as pd
import time

# Python program for implementation of Quicksort
print("\n######################## Quicksort ########################")
# Read Input file
path1 = 'C:\\PGDDS\\BigData\\'
file = '500000SalesRecords.csv'
df = pd.read_csv(path1 + file)
# Get length of list
org_arr = df.OrderID.values
print ("\nLength of list is:") 
numPoints = len(org_arr)
print(numPoints)
print ("\nOriginal List is:") 
print(org_arr)
# Do quick sort
start_time = time.time()
df = df.sort_values(by='OrderID',kind="quicksort")
sort_arr = df['OrderID']
print("\n---Quicksort Execution time: %s seconds ---" % (time.time() - start_time))
print ("\nSorted list is:") 
print(sort_arr)

# Python program for implementation of Mergesort
print("\n######################## Mergesort ########################")
# Read Input file
path1 = 'C:\\PGDDS\\BigData\\'
file = '500000SalesRecords.csv'
df = pd.read_csv(path1 + file)
# Get length of list
org_arr = df.OrderID.values
print ("\nLength of list is:") 
numPoints = len(org_arr)
print(numPoints)
print ("\nOriginal List is:") 
print(org_arr)
# Do merge sort
start_time = time.time()
df = df.sort_values(by='OrderID',kind="mergesort")
sort_arr = df['OrderID']
print("\n---Mergesort Execution time: %s seconds ---" % (time.time() - start_time))
print ("\nSorted list is:") 
print(sort_arr)







 
  


```
